```python
import os
import pandas as pd

def get_CSV_files(file_path):
    csv_files = []  # 保存CSV文件的列表
    for filename in os.listdir(file_path):
        if filename.endswith(".CSV"):
            csv_files.append(os.path.join(file_path, filename))
    return csv_files
```

```python
def get_csv_files(file_path):
    csv_files = []  # 保存CSV文件的列表
    for filename in os.listdir(file_path):
        if filename.endswith(".csv"):
            csv_files.append(os.path.join(file_path, filename))
    return csv_files
```

```python
file_path = './data'
files = get_CSV_files(file_path)
files
```

```python
dic = {'代码':'stock_code','简称':'stock','日期':'trade_date','开盘价(元)':'open','最高价(元)':'high',
       '最低价(元)':'low','收盘价(元)':'close','成交量(股)':'volume','涨跌幅(%)':'ret','均价(元)':'avg','换手率(%)':'TurnoverRate'}
```

```python
for file in files:
    try:
        df = pd.read_csv(file, encoding='utf-8')
    except UnicodeDecodeError:  
        df = pd.read_csv(file, encoding='gbk')

    df.rename(columns=dic, inplace=True)
    df = df.drop(df.columns[-1], axis=1)
    data_folder = './data/data'
    if not os.path.exists(data_folder):
        os.makedirs(data_folder)

    new_file = (file.split('\\')[1]).split('.')[0] + '_processed.csv'
    full_path = os.path.join(data_folder, new_file)

    df.to_csv(full_path, encoding='utf-8-sig', index=False)
```

```python
#遍历每个sheet并按index合并数据

xls = pd.ExcelFile('./data/ind_df.xlsx')
ind_df = pd.DataFrame()
for sheet_name in xls.sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet_name)
    df.columns = df.iloc[1]
    df = df.iloc[2:]
    df = df.melt(id_vars = 'Date',var_name = 'stock_code',value_name = sheet_name)
    df = df.rename(columns = {'Date':'trade_date'})
    
    df['trade_date'] = pd.to_datetime(df['trade_date'])
    
    if ind_df.empty:
        ind_df = df
    else:
        ind_df = pd.merge(ind_df, df, on=['trade_date', 'stock_code'], how='outer')
```

```python
df = pd.read_csv('./data/ind_df.csv')

# 根据 stock_code 分组
grouped = df.groupby('stock_code')

# 遍历每个分组并保存为独立的 CSV 文件
for stock_code, group in grouped:
    group = group.iloc[:,1:]
    filename = f'./data/data/{stock_code}.csv'
    
    # 将分组的数据保存为 CSV 文件
    group.to_csv(filename, index=False)
```
