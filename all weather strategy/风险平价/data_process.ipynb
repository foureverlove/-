```python
import pandas as pd
df = pd.read_csv('原始数据.csv')
df.columns = df.iloc[0]
df = df.iloc[1:]
```

```python
df
```

```python
df_melted = df.melt(id_vars='Date', var_name='ts_code', value_name='pct_chg')
df_melted.rename(columns = {'Date':'trade_date'}, inplace=True)
df_melted['pct_chg']= df_melted['pct_chg'].fillna(0)
df_melted['trade_date'] = pd.to_datetime(df_melted['trade_date'])
df_melted['trade_date'] = df_melted['trade_date'].dt.strftime('%Y%m%d')
df_melted.to_csv('ret_df.csv')
```

```python
bond = pd.read_excel('bond_rate.xlsx')
bond.columns = bond.iloc[0]
bond = bond.iloc[1:]
bond
bond_melted = bond.melt(id_vars='Date',value_name='close')
bond_melted.rename(columns = {'Date':'trade_date','DR007.IB':'close'}, inplace=True)
bond_melted['close']= bond_melted['close'].fillna(bond_melted['close'].mean())
bond_melted['trade_date'] = pd.to_datetime(bond_melted['trade_date'])
bond_melted['trade_date'] = bond_melted['trade_date'].dt.strftime('%Y%m%d')
bond_melted.to_csv('bond_rate.csv')
```
